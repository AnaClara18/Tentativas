```python
import pandas as pd 
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
```

```python
df = pd.read_csv('/home/anaclara/Documents/Tentativas Ana/dados_abertos/dados_abertos___estudantes_da_graduacao_2024-02-15T15_15_39.003984-03_00.csv')
```

```python
df.info()
```

```python
df_alunos = df[df['forma_ingresso'].str.contains('SISU', case=False, na=False)]

display(df_alunos)
```

```python
df.sample(6)
```

```python
df_agrupar = df.groupby('curso').size().reset_index(name='Quantidade')
```

```python
display(df_agrupar)
display(df_agrupar.info())
```

```python
df_encontar = df.loc[df['registro_academico'] == 202325010]

display(df_encontar)
```

```python
df_situacao = df[df['situacao_nome'].str.contains('desistiu do curso', case=False, na=False)]

display(df_situacao)
```

```python
display(df_situacao.sample(10))
```

```python
display(df_situacao.info())
```

```python
df_turno = df[df['curso_turno'].str.contains('notu', case=False, na=False)]

display(df_turno)
```

```python
df[df['titulo_academico'].str.contains('Licenc', case=False, na=False)]
```

```python
df_inicio = df.groupby('semestre_letivo_ingresso').size().reset_index(name='Quantidade')
```
